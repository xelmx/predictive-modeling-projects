This code snippet defines a function called check_outliers to identify potential outliers in the dataset. It utilizes the z-score method, where data points with z-scores exceeding a specified threshold are considered outliers. The function calculates z-scores for all numerical columns and returns the count of outliers per column. This information helps in assessing data quality and determining if any further outlier treatment is necessary.

```python
def check_outliers(usedcars, threshold=3):
    numeric_data = usedcars.select_dtypes(include=['number'])
    z_scores = np.abs((numeric_data - numeric_data.mean()) / numeric_data.std())
    outliers = (z_scores > threshold).sum()
    return outliers

# Check outliers
outliers = check_outliers(df)
print("\nOutliers per column:")
print(outliers)
```

After the data cleaning and preprocessing steps, we proceed with correlation analysis to understand the relationships between different features in the dataset. This analysis helps us identify potential dependencies and associations between variables, providing valuable insights for feature selection and model building.

To conduct correlation analysis, we first create a temporary copy of the DataFrame (df_temp) to avoid modifying the original dataset. Since the dataset contains categorical variables, we need to encode them numerically to calculate correlations. We utilize LabelEncoder from sklearn.preprocessing to temporarily transform categorical columns into numerical representations. This encoding allows us to calculate correlations between all variables, including both numerical and categorical ones.

Once the categorical columns are encoded, we calculate the correlation matrix using the corr() method of the Pandas DataFrame. This matrix quantifies the linear association between pairs of variables, with values ranging from -1 to +1. A positive correlation indicates a direct relationship, where an increase in one variable tends to be associated with an increase in the other. Conversely, a negative correlation suggests an inverse relationship.

To visualize the correlation matrix, we employ a heatmap using the seaborn library. The heatmap provides a clear and intuitive representation of the correlations between variables, where the color intensity reflects the strength and direction of the relationship. We customize the heatmap with annotations to display the correlation values, making it easier to interpret the results.

Furthermore, we extract the top correlated column pairs from the correlation matrix for further investigation. This allows us to focus on the most significant relationships and potentially identify redundant or highly correlated features.

Finally, to maintain the integrity of the DataFrame, we decode the categorical columns back to their original form using the stored LabelEncoder objects. This ensures that the DataFrame remains in its original state after the correlation analysis.

```python
df_temp = df.copy()

label_encoders = {}
for col in df_temp.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_temp[col] = le.fit_transform(df_temp[col])
    label_encoders[col] = le

correlation_matrix = df_temp.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (With Temporarily Encoded Categorical Columns)")
plt.show()

correlation_pairs = correlation_matrix.unstack().reset_index()
correlation_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']

correlation_pairs = correlation_pairs[correlation_pairs['Feature 1'] != correlation_pairs['Feature 2']]

correlation_pairs = correlation_pairs.reindex(correlation_pairs['Correlation'].abs().sort_values(ascending=False).index)

correlation_pairs = correlation_pairs.iloc[::2]

print("\n Top Correlated Column Pairs:")
print(correlation_pairs.head(10))

for col, le in label_encoders.items():
    df_temp[col] = le.inverse_transform(df_temp[col])
```

This code prepares features, applies one-hot encoding, maintains class distribution, trains the BaggingClassifier, and measures performance metrics in order to evaluate a Bagging ensemble using Logistic Regression using 5-fold Stratified Cross-Validation on a student performance dataset.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.ensemble import BaggingClassifier

#Load the modified
mdf = pd.read_csv('/content/modified_student_prediction.csv')

import warnings
warnings.filterwarnings("ignore")

X = mdf.drop(columns=['Grade'])
y = mdf['Grade']

# Encode categorical variables
X = pd.get_dummies(X, drop_first=True)

# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize metric storage
accuracy_list, precision_list, recall_list, f1_list = [], [], [], []

# K-Fold Loop
for train_idx, test_idx in skf.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# Logistic Regression + Bagging

    base_model = LogisticRegression(C=1.0, max_iter=1000, solver='newton-cg')
    model = BaggingClassifier(estimator=base_model, n_estimators=10, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Metrics
    accuracy_list.append(accuracy_score(y_test, y_pred))
    precision_list.append(precision_score(y_test, y_pred, average='weighted', zero_division=0))
    recall_list.append(recall_score(y_test, y_pred, average='weighted', zero_division=0))
    f1_list.append(f1_score(y_test, y_pred, average='weighted', zero_division=0))

# Output average metrics
print("Logistic Regression 5-Fold Results:")
print("Average Accuracy: {:.3f}".format(np.mean(accuracy_list)))
print("Average Precision: {:.3f}".format(np.mean(precision_list)))
print("Average Recall: {:.3f}".format(np.mean(recall_list)))
print("Average F1-score: {:.3f}".format(np.mean(f1_list)))
```

This code uses a Bagging ensemble with Support Vector Machine (SVM) as the base learner and Stratified 5-Fold Cross-Validation to predict student grades. It loads and processes the student dataset, encodes categorical variables, initializes a BaggingClassifier, splits the data into 5 folds, trains the model, predicts on the validation set, and collects evaluation metrics.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC

#Load the modified
mdf = pd.read_csv('/content/modified_student_prediction.csv')

import warnings
warnings.filterwarnings("ignore")

X = mdf.drop(columns=['Grade'])
y = mdf['Grade']

# Encode categorical variables (if not already encoded)
X = pd.get_dummies(X, drop_first=True)

# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize metric storage
accuracy_list, precision_list, recall_list, f1_list = [], [], [], []

# K-Fold Loop
for train_idx, test_idx in skf.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# SVM + Bagging
    base_model = SVC(C=1, kernel='rbf', gamma='scale', probability=True, random_state=42)
    model = BaggingClassifier(estimator=base_model, n_estimators=10, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Metrics
    accuracy_list.append(accuracy_score(y_test, y_pred))
    precision_list.append(precision_score(y_test, y_pred, average='weighted', zero_division=0))
    recall_list.append(recall_score(y_test, y_pred, average='weighted', zero_division=0))
    f1_list.append(f1_score(y_test, y_pred, average='weighted', zero_division=0))

# Output average metrics
print("SVM 5-Fold Results:")
print("Average Accuracy: {:.3f}".format(np.mean(accuracy_list)))
print("Average Precision: {:.3f}".format(np.mean(precision_list)))
print("Average Recall: {:.3f}".format(np.mean(recall_list)))
print("Average F1-score: {:.3f}".format(np.mean(f1_list)))
```

This code uses a stacking ensemble learning method to predict student grades based on various features. Stacking combines the strengths of multiple models (base learners) and uses a meta-learner to make the final decision, ideally improving performance over any single model alone. The code loads the student dataset and prepares it by separating input features (X) from the target variable (grade). Stratified K-Fold cross-validation is used to ensure a similar distribution of target classes. Logistic Regression, Decision Tree, and Support Vector Machine (SVM) are defined as base learners, with SVM wrapped in a pipeline with StandardScaler. The StackingClassifier is defined using these models as base learners, with a Logistic Regression model as the meta-learner. The model is trained on the training data of each fold and predictions are made on the test set. Performance metrics like accuracy, precision, recall, and F1-score are computed for each fold. The script prints the average of each metric across all five folds, providing a summary of the model's expected real-world performance.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Load dataset
mdf = pd.read_csv('/content/modified_student_prediction.csv')
X = mdf.drop(columns=['Grade'])
y = mdf['Grade']

# Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Metric storage
accuracy_list, precision_list, recall_list, f1_list = [], [], [], []

# Base learners
estimators = [
    ('lr', LogisticRegression(max_iter=1000, solver='newton-cg')),
    ('dt', DecisionTreeClassifier(max_depth=15, criterion='entropy', min_samples_split=25)),
    ('svm', make_pipeline(StandardScaler(), SVC(C=1000, kernel='rbf', probability=True)))  # Pipeline to scale for SVM
]

# Meta-learner: Logistic Regression
stacking_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5,
    passthrough=True  # Pass original features along with base predictions to meta-model
)

# K-Fold evaluation
for train_idx, test_idx in skf.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    stacking_model.fit(X_train, y_train)
    y_pred = stacking_model.predict(X_test)

    accuracy_list.append(accuracy_score(y_test, y_pred))
    precision_list.append(precision_score(y_test, y_pred, average='weighted', zero_division=0))
    recall_list.append(recall_score(y_test, y_pred, average='weighted', zero_division=0))
    f1_list.append(f1_score(y_test, y_pred, average='weighted', zero_division=0))

# Print results
print("Stacking Classifier 5-Fold Results:")
print("Average Accuracy: {:.3f}".format(np.mean(accuracy_list)))
print("Average Precision: {:.3f}".format(np.mean(precision_list)))
print("Average Recall: {:.3f}".format(np.mean(recall_list)))
print("Average F1-score: {:.3f}".format(np.mean(f1_list)))
```
